```python
def hate_data(data: list) -> list:
  '''
  Add all hate tweets from the dataset into a list
  Parameters:
    data: a list of multiple tweets
  Returns:
    a list of tweets that are labeled hate
  '''
  hates = []
  for element in data:
    if element[1] == 'hate':
      hates.append(element[2])
  return hates

def non_hate_data(data: list) -> list:
  '''
  Add all non-hate tweets from the dataset into a list
  Parameters:
    data: a list of multiple tweets
  Returns:
    a list of tweets that are not labeled hate
  '''
  not_hates = []
  for element in data:
    if element[1] == 'not-hate':
      not_hates.append(element[2])
  return not_hates


hates = hate_data(results)
not_hates = non_hate_data(results)
```

```python
def binary(nums: list, x: int) -> int:
    '''
      Find the index of an integer in a list
      Parameters:
        nums: A list of integers
        x: the integer to search for
      Returns:
        the index of the number x in the list  
    '''
    start = 0
    end = len(nums) - 1
    while start <= end:
        mid = start + ((end - start) // 2)
        if nums[mid] < x:
            start = mid + 1
        elif nums[mid] > x:
            end = mid - 1
        else:
            return mid
    return start

def k_close(nums: list, x: int, k: int) -> list:
    '''
    Find k values as close as possible to x in a list
    Parameters:
      nums: a list of numbers
      x: target value
      k: number of values that we want to find
    Returns:
      List of k closest values to x
    '''
    no = binary(nums, x)
    lhs = no - 1
    rhs = no
    while k > 0:
        if lhs < 0 or (rhs < len(nums) and abs(nums[lhs] - x) > abs(nums[rhs] - x)):
            rhs = rhs + 1
        else:
            lhs = lhs - 1
        k = k - 1
    return nums[lhs + 1: rhs]
```
